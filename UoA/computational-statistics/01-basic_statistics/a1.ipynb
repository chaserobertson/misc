## Question 1
The data in scorest.txt represent the scores that 50 students earned on a college qualification test. Use R expressions to do the followings.

```r
d <- read.csv('scorest.txt', header=F)
x <- d$V1
```

```r
hist(x)
```

```r
x_mean <- mean(x)
x_sd <- sd(x)
x_median <- median(x)

x_mean
x_sd
x_median
```

```r
# sort the negative count of each distinct value, negate again and return the first value (the mode and its frequency)
-sort(-table(x))[1]
```

(d) The coding of a data set refers to the operation of subtracting (or adding) a constant to each
observation and then dividing (multiplying) by another constant. The coding by subtracting the
sample mean of data and dividing by the sample standard deviation is called standardization and
the coded data is called standardized data. Standardize the data and calculate the mean and the
variance of the standardized data.

```r
x_stan <- (x - x_mean) / x_sd

mean(x_stan)
var(x_stan)
```

## Question 2
A student buys a lottery ticket for $1. For every 1000 tickets sold, 2 bicycles are to be given away in a
drawing.

(a) What is the probability that the student will win a bicycle?

```r
student_tickets <- 1
total_tickets <- 1000
reward_bicycles <- 2

student_win_probability <- student_tickets / (total_tickets / reward_bicycles)
student_win_probability
```

(b) If each bicycle is worth $160, determine the student’s expected gain.

```r
gain_per_win <- 160

expected_gain <- student_win_probability * gain_per_win
paste0("$", expected_gain)
```

3. It is anticipated that a new instructional method will more effectively improve the reading ability of
elementary-school children than the standard method currently in use. To test this conjecture, 16
children are divided at random into two groups of 8 each. One group is instructed using the standard
method and the other group is instructed using the new method. The children’s scores on a reading
test are found to be:  
  
Reading test scores  
Standard | 65 70 76 63 72 71 68 68  
New | 75 80 72 77 69 81 71 78

```r
standard <- c(65, 70, 76, 63, 72, 71, 68, 68)
new <- c(75, 80, 72, 77, 69, 81, 71, 78)
```

(a) Analyze the data and make an inference about the difference of the population mean scores
achieved using the standard and new methods of instruction.

```r
mean(new) - mean(standard)
```

```r
combined <- c(standard, new)
mid <- length(combined)/2

mean_diffs <- 0
for (i in 1:10000) {
    s <- sample(combined)
    s1 <- head(s, mid)
    s2 <- tail(s, mid)
    mean_diffs[i] <- mean(s1) - mean(s2)
}
hist(mean_diffs)
```

The observed mean difference of 6.25 is seen very rarely in randomly selected groups from the same data, as can be seen in the above histogram. The data strongly supports the hypothesis that the new method of reading instruction produced higher test scores.
