# Question 1
An environmental scientist identified a point source for E. Coli at the edge of a stream. She then measured y =E. Coli, in colony forming units per 100 ml water, at different distances, in feet, downstream from the point source. Suppose she obtains the following pairs of (x,y).

| Variable | Observations |
| - | - |
| x | 100 150 250 250 400 650 1000 1600 |
| y | 21 20 24 17 18 10 11 9 |

(a) Transform the x values to x′= log10 x and plot the scatter diagram of y versus x′.

```r
x = c(100, 150, 250, 250, 400, 650, 1000, 1600)
y = c(21, 20, 24, 17, 18, 10, 11, 9)

x_ = log10(x)
plot(x_, y, pch=19)
```

(b) Fit a straight line regression to the transformed data.

```r
fit = lm(y ~ x_)
coef(summary(fit))
```

(c) Obtain a 90% confidence interval for the slope of the regression line.

```r
confint(fit, 'x_', level=0.9)
```

(d) Estimate the expected y value corresponding to x = 300 and give a 95% confidence interval.

```r
pred_x = list(x_=log10(300))
predict(fit, pred_x, interval='confidence', level=0.95)
```

# Question 2
As part of the multilab study, four fabrics are tested for flammability at the National Bureau of Standards. The following burn times in seconds are recorded after a paper tab is ignited on the hem of a dress made of each fabric.

| Fabric 1 | Fabric 2 | Fabric 3 | Fabric 4 |
| - | - | - | - |
| 17.8 | 11.2 | 11.8 | 14.9 |
| 16.2 | 11.4 | 11.0 | 10.8 |
| 17.5 | 15.8 | 10.0 | 12.8 |
| 17.4 | 10.0 | 9.2 | 10.7 |
| 15.0 | 10.4 | 9.2 | 10.7 |

(a) State the statistical model and present the ANOVA table. With α = 0.05, test the null hypothesis of no difference in the degree of flammability for the four fabrics.

> The model is: $y_{i,j} = \mu + \tau_i + \epsilon_{i,j}$ for $i = 1,2,3,4$ (n_fabrics) and $j = 1,2,3,4,5$ (n_tests). $y_{i,j}$ represents each observation, $\mu$ the mean of all observations, $\tau$ the effect of each fabric, and $\epsilon$ random experimental error with distribution N(0,1).

```r
fabrics = data.frame(fabric=as.factor(rep(1:4, each=5)),
                     burn=c(17.8, 16.2, 17.5, 17.4, 15,
                            11.2, 11.4, 15.8, 10, 10.4,
                            11.8, 11, 10, 9.2, 9.2,
                            14.9, 10.8, 12.8, 10.7, 10.7))
fit <- lm(burn ~ fabric, data=fabrics)
anova(fit)
```

> The overall ANOVA p-value is less than the selected $\alpha=0.05$, so we reject the null hypothesis of no difference in the degree of flammability for the four fabrics.

(b) If the null hypothesis is rejected, construct simultaneous confidence intervals to determine the fabric(s) with the lowest mean burn time.

```r
aov <- aov(burn ~ fabric, data=fabrics)
tukey <- TukeyHSD(aov)
round(tukey$fabric, 3)
```

> Fabric 2, 3, and 4 have the lowest mean burn time. Only the interactions with Fabric 1 have differences with high enough confidence to claim significant difference from the others. Those confidence intervals do not include 0, so the available evidence only indicates that Fabric 1 has the highest mean burn time.

(c) Plot the residuals and comment on the plausibility of the assumptions.

```r
plot(aov, which=1)
```

```r
bartlett.test(burn ~ fabric, data=fabrics)
```

> The residual plot suggests that variance may be somewhat different between fabrics. However, Bartlett's test does not indicate significant difference, so it is plausible to assume equality of variance between fabrics.

(d) If the tests had been conducted one at a time on a single mannequin, how would you have randomized the fabrics tested in this experiment?

> I would have randomized the order of each fabric sample's burning on the mannequin. If each fabric's samples were burned in order, the residual heat on the mannequin may have some effect on the results. Randomizing the fabric sample order strengthens the assumption that expectation of the random error term $\epsilon$ is zero.

# Question 3
Many industrial air pollutants adversely affect plants. Sulphur dioxide causes leaf damage in the form of intraveinal bleaching in many sensitive plants. In a study of the effect of a given concentration of sulphur dioxide in the air on three types of garden vegetables, 40 plants of each type are exposed to the pollutant under controlled greenhouse conditions. The frequencies of severe leaf damage are recorded in the following table:

Leaf damage

| Plant | Severe | Moderate or none | Total |
| -:- | -:- | - | -:- |
| Lettuce | 32 | 8 | 40 |
| Spinach | 28 | 12 | 40 |
| Tomato | 19 | 21 | 40 |
| Total | 79 | 41 | 120 |

Analyze these data to determine if the incidence of severe leaf damage is alike for the three types of plants. In particular:

(a) Formulate the null hypothesis.

>With $S$ as the proportion of severe leaf damage in each plant's observations,
$$H_0: S_{lettuce} = S_{spinach} = S_{tomato}$$

(b) Test the null hypothesis with α = 0.05.

```r
leaf <- rbind(c(32, 8), c(28, 12), c(19, 21))
chisq.test(leaf)
```

> Pearson's Chi-squared test estimated a p-value below our threshold of 0.05, so we reject the null hypothesis that severity of leaf damage is alike for the three types of plants.

(c) Construct three individual 98% confidence intervals.

```r
cat(qbinom(0.01, 40, 32/40)/40, " < S_Lettuce <", 
    qbinom(0.99, 40, 32/40)/40, fill=T)
cat(qbinom(0.01, 40, 28/40)/40, "< S_Spinach <", 
    qbinom(0.99, 40, 28/40)/40, fill=T)
cat(qbinom(0.01, 40, 19/40)/40, "  < S_Tomato  <", 
    qbinom(0.99, 40, 19/40)/40, fill=T)
```

# Question 5
An investigation is conducted to determine if the exposure of women to atomic fallout influences the rate of birth defects. A sample of 500 children born to mothers who were exposed to the atomic explosion in Hiroshima is to be studied. A sample of 400 children from a Japanese island far removed from the site of the atomic explosion forms the control group. Suppose that the following data for the incidence of birth defects are obtained:

| Exposure | Defect Present | Defect Absent | Total |
| - | - | - | - |
| Mother exposed | 84 | 416 | 500 | 
| Mother not exposed | 43 | 357 | 400 | 
| Total | 127 | 773 | 900 | 

(a) Do these data indicate that there is a different rate of incidence of birth defects for the two groups of mothers?

```r
mothers <- rbind(c(84, 416), c(43, 357))
prop.test(mothers)
```

> These data **do** indicate a difference in the proportion/rate of birth defects between the two groups of mothers.

(b) Use the normal test for testing the equality of two population proportions with a two-sided alternative. Verify the relation $χ^2 = z^2$ by comparing their numerical values.

> With the normal binomial proportion test, $H_0: \hat{p}_1 - \hat{p}_2$ is approximately normally distributed with $E(\hat{p}_1 - \hat{p}_2) = 0$, so $z = \frac{\hat{p}_1 - \hat{p}_2} {\sqrt{(\hat{p}(1 - \hat{p})} \sqrt{\frac{1}{n_1} + \frac{1}{n_2}}}$

```r
n1 = 500
n2 = 400
phat1 = 84 / n1
phat2 = 43 / n2
phat = (84+43) / (n1+n2)

z = (phat1 - phat2) / (sqrt(phat*(1-phat)) * sqrt((1/n1 + 1/n2)))
cat('z^2:', z^2, fill=T)
round(prop.test(mothers, correct=F)$statistic, 6)
```

> As illustrated above, $\chi^2 = z^2$.
